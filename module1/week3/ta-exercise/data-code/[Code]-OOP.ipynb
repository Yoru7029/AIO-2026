## 1. Class & Object
`class` là khuôn mẫu; `object` là thực thể cụ thể. `__init__` là hàm khởi tạo, `self` trỏ tới chính object.

```python
class Dog:
    def __init__(self, name: str, age: int) -> None:
        self.name = name      # thuoc tinh (attribute)
        self.age = age

    def eat(self) -> str:     # phuong thuc (method)
        return 'Chicken' if self.age <= 1 else 'Fish'


dog_1 = Dog(name="Chow Chow", age=2)
print(dog_1.name)    # Chow Chow
print(dog_1.eat())   # Fish
```

## 2. Access modifiers: public / protected / private
- `owner` → **public**: truy cập tự do.
- `_bank` → **protected**: quy ước chỉ dùng nội bộ/lớp con.
- `__pin` → **private**: Python *name-mangling*, truy cập trực tiếp sẽ lỗi `AttributeError`.

```python
class Account:
    def __init__(self, owner: str, bank: str, pin: str) -> None:
        self.owner = owner    # public
        self._bank = bank     # protected (quy uoc)
        self.__pin = pin      # private (name-mangling)


acc = Account("Mai", "AIO Bank", "1234")
print(acc.owner)    # OK  -> Mai
print(acc._bank)    # OK  -> AIO Bank

try:
    print(acc.__pin)            # truy cap private -> loi
except AttributeError as e:
    print("AttributeError:", e)
```

## 3. Encapsulation — getter / setter
Với thuộc tính private, truy cập gián tiếp qua getter/setter. Đây là tinh thần **đóng gói**: che giấu dữ liệu, chỉ thao tác qua giao diện công khai.

```python
class Profile:
    def __init__(self, name: str) -> None:
        self.__name = name

    def get_name(self) -> str:
        return self.__name

    def set_name(self, name: str) -> None:
        self.__name = name


p = Profile("Chow Chow")
print(p.get_name())   # Chow Chow
p.set_name("Chaw Chaw")
print(p.get_name())   # Chaw Chaw
```

## 4. Abstraction — Abstract class
`abstractmethod` bắt buộc lớp con phải hiện thực method; không thể tạo object trực tiếp từ abstract class.

```python
from abc import ABC, abstractmethod


class Shape(ABC):
    @abstractmethod
    def compute_area(self) -> float:
        pass


class Square(Shape):
    def __init__(self, side: float) -> None:
        self.__side = side

    def compute_area(self) -> float:
        return self.__side * self.__side


print(Square(5).compute_area())   # 25
```

## 5. Inheritance — kế thừa & super()
Lớp con kế thừa lớp cha, dùng `super()` để tái sử dụng phần khởi tạo/hành vi của cha.

```python
class Employee:
    def __init__(self, name: str, salary: float) -> None:
        self._name = name
        self._salary = salary

    def compute_salary(self) -> float:
        return self._salary


class Manager(Employee):
    def __init__(self, name: str, salary: float, bonus: float) -> None:
        super().__init__(name, salary)
        self.__bonus = bonus

    def compute_salary(self) -> float:
        return super().compute_salary() + self.__bonus


print(Manager("Mai", 100, 50).compute_salary())   # 150
```

## 6. Polymorphism — đa hình
Cùng tên method `show()` nhưng lớp con ghi đè (override) để có hành vi khác nhau.

```python
class A:
    def __init__(self, num: int) -> None:
        self.num = num

    def show(self) -> None:
        print(self.num)


class B(A):
    def show(self) -> None:        # override
        print(self.num * self.num)


A(3).show()   # 3
B(3).show()   # 9
```

## 7. Magic methods (dunder methods)
Magic method (còn gọi là dunder method) là các method có tên dạng `__ten__` được Python gọi tự động trong một số tình huống nhất định. Nhờ đó, object do ta định nghĩa có thể dùng được với các cú pháp có sẵn của Python. Một số method thường gặp:
- `__str__(self)`: trả về chuỗi mô tả object dành cho người đọc, được gọi khi thực hiện `print(obj)` hoặc `str(obj)`.
- `__repr__(self)`: trả về chuỗi biểu diễn object dành cho lập trình viên (thường đủ thông tin để tái tạo object), được gọi khi hiển thị object trong danh sách hoặc trên môi trường tương tác, hoặc khi gọi `repr(obj)`.
- `__len__(self)`: trả về số phần tử của object, cho phép gọi `len(obj)`.
- `__lt__(self, other)`: định nghĩa toán tử so sánh nhỏ hơn (`<`), nhờ đó `sorted()` và `list.sort()` có thể sắp xếp các object.

```python
class Book:
    def __init__(self, title: str, pages: int) -> None:
        self.title = title
        self.pages = pages

    def __str__(self) -> str:
        return f"'{self.title}' ({self.pages} trang)"

    def __repr__(self) -> str:
        return f"Book(title={self.title!r}, pages={self.pages})"

    def __len__(self) -> int:
        return self.pages

    def __lt__(self, other: "Book") -> bool:
        return self.pages < other.pages


b1 = Book("Python", 320)
b2 = Book("AI", 150)
print(b1)                  # 'Python' (320 trang)        <- __str__
print([b1, b2])            # [Book(...), Book(...)]       <- __repr__
print(len(b1))             # 320                          <- __len__
print(sorted([b1, b2]))    # sap xep theo so trang        <- __lt__
```
