## Bài 1. Xây dựng class tính Sigmoid
`Sigmoid(x) = 1 / (1 + e^{-x})`. Kế thừa `nn.Module`, ghi đè `forward`.
> Lưu ý: input phải là tensor **float**.

```python
import torch
import torch.nn as nn


class MySigmoid(nn.Module):
    def __init__(self) -> None:
        super().__init__()

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        ### Your Code Here
        if not isinstance(x, torch.Tensor):
            raise TypeError("Must be torch.Tensor class")
        
        x = x.float()
        sigmoid = 1 / (1 + torch.exp(-x))
        
        return sigmoid
        ### End Code Here


my_sigmoid = MySigmoid()
x = torch.tensor([1.0, 5.0, -4.0, 3.0, -2.0])
print(my_sigmoid(x))
# Mong doi: tensor([0.7311, 0.9933, 0.0180, 0.9526, 0.1192])
```

## Bài 2. Xây dựng Ward (Person / Student / Teacher / Doctor)
Hoàn thiện `__init__` + `describe` cho 3 lớp con, rồi hoàn thiện các method của `Ward`.

```python
from abc import ABC, abstractmethod


class Person(ABC):
    def __init__(self, name: str, yob: int) -> None:
        self._name = name
        self._yob = yob

    def get_yob(self) -> int:
        return self._yob

    def __lt__(self, other: "Person") -> bool:   # da co san: so sanh theo yob
        return self._yob < other._yob

    @abstractmethod
    def describe(self) -> None:
        pass


class Student(Person):
    def __init__(self, name: str, yob: int, grade: str) -> None:
        ### Your Code Here
        super().__init__(name, yob)
        self._grade = grade
        ### End Code Here

    def __str__(self) -> str:
        ### Your Code Here
        return f"name: {self._name} year of birth: {self._yob} grade: {self._grade}"
        ### End Code Here

    def describe(self) -> None:
        print(self)            # dung __str__


class Teacher(Person):
    def __init__(self, name: str, yob: int, subject: str) -> None:
        ### Your Code Here
        super().__init__(name, yob)
        self._subject = subject
        ### End Code Here

    def __str__(self) -> str:
        ### Your Code Here
        return f"name: {self._name} year of birth: {self._yob} subject: {self._subject}"
        ### End Code Here

    def describe(self) -> None:
        print(self)


class Doctor(Person):
    def __init__(self, name: str, yob: int, specialist: str) -> None:
        ### Your Code Here
        super().__init__(name, yob)
        self._specialist = specialist
        ### End Code Here

    def __str__(self) -> str:
        ### Your Code Here
        return f"name: {self._name} year of birth: {self._yob} specialist: {self._specialist}"
        ### End Code Here

    def describe(self) -> None:
        print(self)
```

```python
class Ward:
    def __init__(self, name: str) -> None:
        self.__name = name
        self.__list_people: list[Person] = list()

    def add_person(self, person: Person) -> None:
        self.__list_people.append(person)

    def describe(self) -> None:
        ### Your Code Here
        print(f"Ward Name: {self.__name}")
        for person in self.__list_people:
            # If the person is a studen
            if isinstance(person, Student):
                print(f"Student - Name: {person._name} - YoB: {person._yob} - Grade: {person._grade}")
            # If the person is a teacher
            elif isinstance(person, Teacher):
                print(f"Teacher - Name: {person._name} - YoB: {person._yob} - Subject: {person._subject}")
            # If the person is a doctor
            elif isinstance(person, Doctor):
                print(f"Doctor - Name: {person._name} - YoB: {person._yob} - Specialist: {person._specialist}")
        ### End Code Here

    def count_doctor(self) -> int:
        ### Your Code Here
        return len([person for person in self.__list_people
                    if isinstance(person, Doctor)])
        ### End Code Here

    def sort_age(self) -> None:
        ### Your Code Here
        return self.__list_people.sort(key=lambda person: person._yob, reverse=True)
        ### End Code Here

    def average_teacher_yob(self) -> float:
        ### Your Code Here
        teacher_yob = [person._yob for person in self.__list_people
                                   if isinstance(person, Teacher)]
        avg_yob = sum(teacher_yob) / len(teacher_yob)
        return avg_yob
        ### End Code Here
```

```python
student1 = Student(name="studentA", yob=2010, grade="7")
teacher1 = Teacher(name="teacherA", yob=1969, subject="Math")
teacher2 = Teacher(name="teacherB", yob=1995, subject="History")
doctor1 = Doctor(name="doctorA", yob=1945, specialist="Endocrinologists")
doctor2 = Doctor(name="doctorB", yob=1975, specialist="Cardiologists")

ward1 = Ward(name="Ward1")
for person in [student1, teacher1, teacher2, doctor1, doctor2]:
    ward1.add_person(person)
ward1.describe()

print(f"\nNumber of doctors: {ward1.count_doctor()}")

print("\nAfter sorting Age of Ward1 people")
ward1.sort_age()
ward1.describe()

print(f"\nAverage year of birth (teachers): {ward1.average_teacher_yob()}")
```

## Bài 3. Xây dựng class Stack (LIFO)
`__init__` và `is_full` đã có sẵn làm mẫu. Hoàn thiện `is_empty`, `push`, `pop`, `top`.

```python
from typing import Optional


class MyStack:
    def __init__(self, capacity: int) -> None:
        self.__capacity = capacity
        self.__stack: list[int] = []

    def is_full(self) -> bool:
        return len(self.__stack) == self.__capacity

    def is_empty(self) -> bool:
        ### Your Code Here
        return len(self.__stack) == 0
        ### End Code Here

    def push(self, value: int) -> None:
        ### Your Code Here
        if self.is_full(): raise Exception("The stack is full")
        self.__stack.append(value)
        ### End Code Here

    def pop(self) -> int:
        ### Your Code Here
        if self.is_empty(): raise Exception("The stack is empty")
        return self.__stack.pop()
        ### End Code Here

    def top(self) -> Optional[int]:
        ### Your Code Here
        return self.__stack[-1] if not self.is_empty() else None
        ### End Code Here

    def __len__(self) -> int:
        ### Your Code Here
        return len(self.__stack)
        ### End Code Here

    def __repr__(self) -> str:
        ### Your Code Here
        return f"Stack: {self.__stack}"
        ### End Code Here
```

```python
stack1 = MyStack(capacity=5)
stack1.push(1)
stack1.push(2)

print(stack1.is_full())   # False
print(stack1.top())       # 2
print(stack1.pop())       # 2
print(stack1.top())       # 1
print(stack1.pop())       # 1
print(stack1.is_empty())  # True
```

## Bài 4. Xây dựng class Queue (FIFO)
`__init__` và `is_full` đã có sẵn làm mẫu. Hoàn thiện `is_empty`, `enqueue`, `dequeue`, `front`.

```python
from typing import Optional


class MyQueue:
    def __init__(self, capacity: int) -> None:
        self.__capacity = capacity
        self.__queue: list[int] = []

    def is_full(self) -> bool:
        return len(self.__queue) == self.__capacity

    def is_empty(self) -> bool:
        ### Your Code Here
        return len(self.__queue) == 0
        ### End Code Here

    def enqueue(self, value: int) -> None:
        ### Your Code Here
        if self.is_full(): raise Exception("The queue is full")
        self.__queue.append(value)
        ### End Code Here

    def dequeue(self) -> int:
        ### Your Code Here
        if self.is_empty(): raise Exception("The queue is empty")
        return self.__queue.pop(0)
        ### End Code Here

    def front(self) -> Optional[int]:
        ### Your Code Here
        return self.__queue[0] if not self.is_empty() else None
        ### End Code Here

    def __len__(self) -> int:
        ### Your Code Here
        return len(self.__queue)
        ### End Code Here

    def __repr__(self) -> str:
        ### Your Code Here
        return f"Queue: {self.__queue}"
        ### End Code Here
```

```python
queue1 = MyQueue(capacity=5)
queue1.enqueue(1)
queue1.enqueue(2)

print(queue1.is_full())   # False
print(queue1.front())     # 1
print(queue1.dequeue())   # 1
print(queue1.front())     # 2
print(queue1.dequeue())   # 2
print(queue1.is_empty())  # True
```

Vận dụng Stack/Queue đã viết để giải 2 bài ứng dụng nhỏ.

```python
# Thu thach 1 (Stack): kiem tra ngoac can bang () [] {}
def is_balanced(text: str) -> bool:
    pairs = {")": "(", "]": "[", "}": "{"}
    stack = MyStack(capacity=len(text) + 1)
    ### Your Code Here
    # Create the list of open bracket
    open_brackets = list(pairs.values())

    # Loop through reversed text
    for character in text[::-1]:
        # Push the corresponding pair to stack if found a closed bracket
        if character in pairs:
            stack.push(open_brackets.index(pairs[character]))
        
        # Pop the stack if found the open pair of the innermost bracket
        elif character in open_brackets and stack.top() == open_brackets.index(character):
            stack.pop()
        
        # Skip if it's a letter
        else: continue
    
    # The text is considered "balanced" if the stack is empty after the loop
    return stack.is_empty()
    ### End Code Here

print(is_balanced("(a[b]{c})"))   # mong doi: True
print(is_balanced("(a[b)c]"))     # mong doi: False
```

```python
# Thu thach 2 (Queue): mo phong hang doi may in theo thu tu FIFO
def printer_simulation(jobs: list[str]) -> None:
    queue = MyQueue(capacity=len(jobs) + 1)
    ### Your Code Here
    for idx in range(len(jobs)):
        queue.enqueue(idx)
    
    while not queue.is_empty():
        idx = queue.dequeue()
        print(jobs[idx])

    ### End Code Here

printer_simulation(["BaiTap1.pdf", "Slide.pptx", "Anh.png"])
```
